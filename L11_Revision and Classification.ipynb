```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.datasets import load_iris
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# loading the IRIS dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
iris.target_names
```

### Multinomial Logistic Regression

```python
X_multi = df[['sepal length (cm)','sepal width (cm)']].values.reshape(150,2)
y_multi = df['target'].values
```

```python
# Split the data into training/testing sets (80:20 split)
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size= 0.20, random_state= 101)
```

```python
# default regularization is l2 with alpha of 1, by putting c = 30 (inverse of alpha)
#I am decreaing the regularization coefficient
from sklearn.linear_model import LogisticRegression
log_reg_multi = LogisticRegression(C=30,random_state=42) # try 10, 0.1 and 0.01
log_reg_multi.fit(X_train_multi, y_train_multi)
```

##### Plotting the decision boundary

```python
x_index = 0
y_index = 1
```

```python
x0, x1 = np.meshgrid(
    np.linspace(iris.data[:, x_index].min(), iris.data[:, x_index].max(), 100).reshape(-1, 1),
    np.linspace(iris.data[:, y_index].min(), iris.data[:, y_index].max(), 100).reshape(-1, 1)
)

X_new = np.c_[x0.ravel(), x1.ravel()]
y_pred = log_reg_multi.predict(X_new)
zz = y_pred.reshape(x0.shape)

plt.figure(figsize=(10, 5))
plt.contourf(x0, x1, zz, cmap=plt.cm.Spectral)
plt.scatter(X_multi[:,0], X_multi[:,1], c=y_multi)

plt.xlabel('sepal length')
plt.ylabel('sepal width')
```

### Checking Accuracy

```python
print("training set score is : {:.3f}" .format(log_reg_multi.score(X_train_multi,y_train_multi)))
print("test set score is : {:.3f}" .format(log_reg_multi.score(X_test_multi,y_test_multi)))
```

### Confusion Matrix

```python
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_test_predictions = log_reg_multi.predict(X_test_multi)
ConfusionMatrixDisplay.from_predictions(y_test_multi,y_test_predictions)
```

### Precision, Recall and f1-score

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
precision = precision_score(y_test_multi,y_test_predictions, average='weighted')
```

```python
recall = recall_score(y_test_multi,y_test_predictions, average='weighted')
```

```python
f1_score_value = f1_score(y_test_multi,y_test_predictions, average='weighted')
```
